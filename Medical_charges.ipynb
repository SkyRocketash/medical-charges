```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#file name:medical-charges.csv
df=pd.read_csv("medical-charges.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x=df.drop(["charges"],axis=1)
x.head()
```

```python
y=df["charges"]
y.head()
```

```python
x=pd.get_dummies(x,columns=["region","children"])

x.head()
```

```python
x["sex"]=x["sex"].replace({"male":0,"female":1})
x["smoker"]=x["smoker"].replace({"yes":1,"no":0})
x.iloc[:,4:]=x.iloc[:,4:].astype(int)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.svm import SVR
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
models={
    "Decision_Tree":DecisionTreeRegressor(),
    "RandomForest":RandomForestRegressor(),
    "LinearRegression":LinearRegression(),
    "Ridge":Ridge(),
    "Lasso":Lasso(),
    "svr":SVR()
}
for i in range(len(list(models))):
  model=list(models.values())[i]
  model.fit(x_train_scaled,y_train)
  y_train_pred=model.predict(x_train_scaled)
  y_test_pred=model.predict(x_test_scaled)

  #training performance
  train_r2 = r2_score(y_train, y_train_pred)
  train_mse = mean_squared_error(y_train, y_train_pred)
  train_mae = mean_absolute_error(y_train, y_train_pred)

  # Test performance
  test_r2 = r2_score(y_test, y_test_pred)
  test_mse = mean_squared_error(y_test, y_test_pred)
  test_mae = mean_absolute_error(y_test, y_test_pred)

  print(list(models.keys())[i])
  print("              Train      Test")
  print(f"R Score     : {train_r2:.4f}   {test_r2:.4f}")
  print(f"MSE          : {train_mse:.4f}   {test_mse:.4f}")
  print(f"MAE          : {train_mae:.4f}   {test_mae:.4f}")
  print("=" * 35)
  print("\n")



```
